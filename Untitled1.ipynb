```python
import numpy as np
import os
import re


def from_file_to_tick_volume(file):
    data = pd.read_csv("data/tick/" + file, 
                   header=None, 
                   names=["Symbol", "Date_time", "Ask", "Bid"])
    
    data.index = pd.to_datetime(data["Date_time"], format="%Y%m%d %H:%M:%S.%f")
    data_ohlc = data["Ask"].resample("15Min").count()
    return data_ohlc



tickers = np.unique([file_name[: 6] for file_name in os.listdir('data/tick')])

for ticker in tickers:
    pattern = re.compile(ticker + '*')
    files = list(filter(pattern.search, os.listdir("data/tick")))
    data = pd.concat([from_file_to_tick_volume(file) for file in np.sort(files)])
    pd.DataFrame.to_csv(data, 'data/tick_volume/' + ticker + '.csv')
```

```python
from random import sample

n = 1000
x = np.linspace(0, 4, n)
y = np.sin(x) + np.random.normal(scale=0.001, size=n)

train_idx = sample(range(n), k=int(0.3 * n))

X_train = x[train_idx]
y_train = y[train_idx]
```

```python
from sklearn.linear_model import LinearRegression

preds = []
for k in range(10):
    train_idx = sample(range(n), k=int(0.3 * n))

    X_train = x[train_idx]
    X_train = np.array([X_train, X_train ** 2]).reshape(-1, 2)
    y_train = y[train_idx]
    
    reg = LinearRegression()
    reg.fit(X_train, y_train)

    pred = reg.predict(np.array([x, x ** 2]).reshape(-1, 2))
    preds.append(pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
[plt.plot(x, pred) for pred in preds]
plt.show()
```
